```python
import librosa
import matplotlib.pyplot as plt
import librosa.display
import sklearn

audio_path = 'C:/Users/19144/OneDrive/Desktop/Ice and Fire - King Canyon.mp3'
x, sr = librosa.load(audio_path, sr=None)
```

```python
print(type(x), type(sr))
```

```python
#Let us display the spectogram here
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(12,5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
#Understand zero crossing rate 
x, sr = librosa.load(audio_path)
plt.figure(figsize=(12,5))
librosa.display.waveshow(x, sr=sr)
```

```python
#Understanding zooming in 
n0 = 9000
n1 = 9100
plt.figure(figsize=(12,5))
plt.plot(x[n0:n1])
plt.grid
```

```python
zero_crossings = librosa.zero_crossings(x[n0:n1],pad=False)
print(sum(zero_crossings))
#13 zero crossings
```

```python
#Understanding the spectral centroid...based on frequencies
spectral_centroids = librosa.feature.spectral_centroid(x, sr=sr)[0]
spectral_centroids.shape

frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)

def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)

librosa.display.waveshow(x, sr=sr, alpha=0.4)
plt.plot(t, normalize(spectral_centroids), color='r')
#Wave form has been plotted...there are consistent high-frequencies...
```

```python
#Understand spectral roll-off - Where spectral energy lies...
spectral_rolloff = librosa.feature.spectral_rolloff(x, sr=sr)[0]
librosa.display.waveshow(x, sr=sr, alpha=0.4)
plt.plot(t, normalize(spectral_rolloff), color='r')
```

```python
#Understand MFCC
mfccs = librosa.feature.mfcc(x, sr=sr)
print(mfccs.shape)

librosa.display.specshow(mfccs, sr=sr, x_axis='time')
```
